## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
#     return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)    
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Pipeline: 

**Step1:** set all global parameter used in following function.

In lane_detector function, 

**Step2:** get the shape of image, and calculate the vertices for roi masking.

**Step3:** convert the color space of image from RGB to GRAYSCALE.

**Step4:** do guassian blur to reduce noise

**Step5:** call canny to extract edges from grayscale image

**Step6:** split roi out of edges

**Step7:** draw line on roi

**Step8:** use cv2.addWeighted method to linearly blend roi_with_line and original_img with RGB color space

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.

# the following parameters in global scope are already tuned during the speech course

# use in Gaussian Blur algorithm
kernel_size = 5

# use in Canny algorithm
low_threshold = 50
high_threshold = 150

# use in Hough transform
rho = 2
theta = np.pi/180
threshold = 40
min_line_length = 20
max_line_gap = 30

# Be care for that, image.shape and vertices should be placed in lane_detector, 
# in order that the vision of car (roi region) can dynamically change with different resolution.
# 
# for example. the "chanllenge.mp4" video is 1280*720, which is different from previous two videos.

# img_shape = image.shape
# vertices = np.array([[(0,img_shape[0]),(465, 320), (495, 320), (img_shape[1],img_shape[0])]], dtype=np.int32)

def lane_detector_for_test_image(img, dst_path=""):
    
    img_shape = img.shape
    roi_y_top = img_shape[0]*0.6
    mid_x = img_shape[1]/2         # midden position of x coordinate in the given image
    offset = 10
    vertices = np.array([[(0,img_shape[0]),(mid_x - offset, roi_y_top), (mid_x + offset, roi_y_top), (img_shape[1],img_shape[0])]], dtype=np.int32)
    
    # Originally I thought that extract roi before every matrix manipulation may reduce the volume of calculation.
    # However, if we extract roi before doing canny, the canny will detect the edge of roi range.
    # 
#     roi = region_of_interest(img, vertices)
#     gray_roi = grayscale(roi)
#     blur = gaussian_blur(gray_roi, kernel_size)
#     edges = canny(blur, low_threshold, high_threshold)
#     edges_with_lines = hough_lines(edges, rho, theta, threshold, min_line_length, max_line_gap)
    
    # doing blur explicitly before canny usually get the good effect.
    gray_img = gray_roi =  grayscale(img)
    blur = gaussian_blur(gray_img, kernel_size)
    edges = canny(blur, low_threshold, high_threshold)
    roi = region_of_interest(edges, vertices)
    edges_with_lines = hough_lines(roi, rho, theta, threshold, min_line_length, max_line_gap)
    
    result = weighted_img(edges_with_lines, img)
    
    if dst_path != "":
        plt.imsave(dst_path, result)
    
    return result

image = mpimg.imread('test_images/solidWhiteRight.jpg')
final_img = lane_detector_for_test_image(image)

# for debug
#
plt.imshow(final_img, cmap='gray')
plt.show()

for img_name in os.listdir("test_images/"):
    image = mpimg.imread('test_images/' + img_name)
    lane_detector_for_test_image(image, 'test_images_output/' + img_name)
````

## My Improve method:

**Note: It really helpful to generate roi with both line image and edge to get the idea of what vision our computer really see. This gives me the idea of how to debug on challenge.mp4**

---    
**In process_image function:**

(1) Replace grayscale function with color_select function 
   which restrict computer only see particular like yellow or white. (Tricky Method)
       
   It is important when challenge on challenge.mp4, which has the big noise produced by the scenery with sunlight and shadow intertwinded each other.
       
   **It reduce lot of noise in such situation.**
   
---   
       
**In new draw_lines function:**
    
(1) (Tricky Method) I only keep the each of two points if the slope between them is larger then 30 degree.
    or if the slope between then is lower then -30 degree.
    
   **This trick dramatically improve the stablity of two lane lines.**
    
    
(2) To calculate average slope and intercept, 
    I seperate x and y coordinate from points which belong to positive slope line and negative slope line.
       
    Then use np.polyfit taught in class to get slope and intercept.
       
       
(3) To draw full line with no broken point, I select two y coordicate to reversely calculate corespond x coordinate.
    
    two y coordinate are: 
      
      (a) first y is at the botton of image 
      (b) second y times 0.6 from first y

---
**Note: Both color_select and slope boundary will remove some points which was originally accept in previous method, so be sure to check if there is no point to draw line **

````python
def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # First, we seperate points which belong to positive slope line or negative slope line
    point_in_positive_slope = []
    point_in_negative_slope = []

    # manually select the slope boundary (tricky method)
    positive_degree_bound = 30 / 180 * np.pi
    negative_degree_bound = (-30) / 180 * np.pi
    
    if lines is not None:
        for line in lines:
            for x1,y1,x2,y2 in line:
                slope = (y2-y1) / (x2-x1)
                if slope > positive_degree_bound:
                    point_in_positive_slope.append([x1, y1])
                    point_in_positive_slope.append([x2, y2])
                elif slope < negative_degree_bound:
                    point_in_negative_slope.append([x1, y1])
                    point_in_negative_slope.append([x2, y2])
    
    # seperate x point and y point into two lists, in order to do polyfit to calculate average slope and intercept
    px = [point_in_positive_slope[i][0] for i in range(len(point_in_positive_slope))]    
    py = [point_in_positive_slope[i][1] for i in range(len(point_in_positive_slope))]
    
    nx = [point_in_negative_slope[i][0] for i in range(len(point_in_negative_slope))]    
    ny = [point_in_negative_slope[i][1] for i in range(len(point_in_negative_slope))]    
     
    if len(px) and len(py):
        positive_slope, positive_intercept = np.polyfit(px, py, 1)
        
        # for debug
        # print("positive: (", positive_slope, ", " , positive_intercept, ")")
        pm = positive_slope
        pb = positive_intercept
        
        # IMPORTANT: to draw full line with no broken point, I select two y coordicate to reversely calculate corespond x coordinate.
        # two y coordinate are: (1) first y is at the botton of image (2) second y times 0.6 from first y
        px1 = int(((img.shape)[0] * 0.6 - pb) / pm)
        px2 = int(((img.shape)[0] - pb) / pm)
        py1 = int(pm * px1 + pb)
        py2 = int(pm * px2 + pb)        
        cv2.line(img, (px1, py1), (px2, py2), color, thickness)

        
    if len(nx) and len(ny):    
        negative_slope, negative_intercept = np.polyfit(nx, ny, 1)
        # print("negative: (", negative_slope, ", " , negative_intercept, ")")
        nm = negative_slope
        nb = negative_intercept

        nx1 = int(((img.shape)[0] - nb) / nm)
        nx2 = int(((img.shape)[0] * 0.6 - nb) / nm)
        ny1 = int(nm * nx1 + nb)
        ny2 = int(nm * nx2 + nb)
        cv2.line(img, (nx1, ny1), (nx2, ny2), color, thickness)
````
